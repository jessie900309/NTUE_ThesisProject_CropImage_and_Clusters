## 用多項式係數作為訓練資料

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def fit_polynomial(x_values, y_values, degree):
    coefficients = np.polyfit(x_values, y_values, degree)
    return coefficients

def plot_polynomial(coefficients, x_values, y_values):
    # Generate points for the polynomial curve
    x_curve = np.linspace(min(x_values), max(x_values), 600)
    y_curve = np.polyval(coefficients, x_curve)
    # Plot the original data points
    plt.scatter(x_values, y_values, color='blue', label='Data Points')
    # Plot the polynomial curve
    plt.plot(x_curve, y_curve, color='red', label='Polynomial Fit (Degree {})'.format(len(coefficients)-1))
    # Add labels and legend
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.title('Polynomial Fit')
    plt.legend()
    # Show the plot
    plt.show()
```

### Example

```python
file_path = 'C:/Users/zombie/PycharmProjects/NTUE_ThesisProject_ImageProcessing/datasets/frames_row_data_Ruifang/0_3_0_C0185_mask_area.csv'
df = pd.read_csv(file_path)
x_values = np.array(range(600))
y_values = ((df.iloc[:, -3:-2].values).astype('float32')).reshape(-1)
```

```python
# Fit a polynomial of degree 10 to the data
coefficients = fit_polynomial(x_values, y_values, 10)
# Plot the polynomial curve
plot_polynomial(coefficients, x_values, y_values)
# Print the coefficients
print("Coefficients of the polynomial fit:")
print(coefficients)
```

### Data

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics.cluster import adjusted_rand_score
```

```python

# 只取Red計算係數
coefficients_data = []
folder_path = 'C:/Users/zombie/PycharmProjects/NTUE_ThesisProject_ImageProcessing/datasets/frames_row_data_Ruifang'
file_names = os.listdir(folder_path)
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)
    x_values = np.array(range(600))
    y_values = ((df.iloc[:, -3:-2].values).astype('float32')).reshape(-1)
    # Fit a polynomial of degree 10 to the data
    coefficients = fit_polynomial(x_values, y_values, 10)
    coefficients_data.append(coefficients)

X = coefficients_data
csv_data = pd.read_csv('data_csv/data_TPH_Ruifang_54stu.csv')
y_true = csv_data["age"].values.tolist()
```

```python
silhouette_avg = []
for i in range(2,20):
    kmeans_fit = KMeans(n_clusters=i, max_iter=1000).fit(X)
    silhouette_avg.append(silhouette_score(X, kmeans_fit.labels_))
plt.plot(range(2,20), silhouette_avg)
```

```python
k_means = KMeans(n_clusters=4, max_iter=1000)
k_means.fit(X) # 訓練
y_kmeans = k_means.predict(X) # 預測
```

```python
adjusted_rand_score(y_true, y_kmeans)
```

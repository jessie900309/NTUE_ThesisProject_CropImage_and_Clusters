```python
from sklearn.cluster import KMeans
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data：有序的三維數列
data = []
folder_path = 'C:/Users/zombie/PycharmProjects/NTUE_ThesisProject_ImageProcessing/datasets/frames_row_data_Ruifang'
file_names = os.listdir(folder_path)
for file_name in file_names:
    # 完整路徑
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path)
    # 只取面積
    selected_data = df.iloc[:, -3:].values
    data.append(selected_data)

# 轉numpy array
data = np.array(data)

# 三維轉二維
reshaped_data = data.reshape((len(file_names), -1))
```

```python
# K-means K=4
kmeans = KMeans(n_clusters=4)
kmeans.fit(reshaped_data)
# 分群結果
labels = kmeans.labels_
```

### 分群結果

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
for i in range(len(file_names)):
    if labels[i] == 0:
        ax.plot(data[i][:, 0], data[i][:, 1], data[i][:, 2], c='r', alpha=0.5)
    elif labels[i] == 1:
        ax.plot(data[i][:, 0], data[i][:, 1], data[i][:, 2], c='g', alpha=0.5)
    elif labels[i] == 2:
        ax.plot(data[i][:, 0], data[i][:, 1], data[i][:, 2], c='b', alpha=0.5)
    else:
        ax.plot(data[i][:, 0], data[i][:, 1], data[i][:, 2], c='m', alpha=0.5)
ax.set_title('Clustered Data')
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Brown')
plt.grid(True)
plt.show()
```

```python
y_true = []
for f in file_names:
    y_true.append(int(f[2]))
print(len(y_true), " ", len(labels))
```

```python
def show_result(y_true, y_predict):
    for i in range(len(y_true)):
        if labels[i] == 0:
            print(y_true[i], "  ", y_predict[i])
    print("-----")
    for i in range(len(y_true)):
        if labels[i] == 1:
            print(y_true[i], "  ", y_predict[i])
    print("-----")
    for i in range(len(y_true)):
        if labels[i] == 2:
            print(y_true[i], "  ", y_predict[i])
    print("-----")
    for i in range(len(y_true)):
        if labels[i] == 3:
            print(y_true[i], "  ", y_predict[i])
```

```python
show_result(y_true=y_true, y_predict=labels)
```
